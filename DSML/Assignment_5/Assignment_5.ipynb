```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read csv file
cust = pd.read_csv("/content/Customers - Customers.csv")
cust.head()
```

```python
# Count plot of count of each profession
plt.figure(figsize = (10,5))
sns.countplot(x = "Profession", data = cust, palette = "flare")
plt.title('Profession Counts')
plt.xlabel('Profession')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot
plt.figure(figsize = (10,6))
col1 = "Age"
col2 = "Work Experience"
plt.scatter(cust[col1], cust[col2], c = "blue", alpha = 0.5)

plt.xlabel("Age")
plt.ylabel("Work Experience")
plt.title("Age vs Work Experience")
plt.show()
```

```python
#Hist plot
plt.figure(figsize=(10, 5))
sns.histplot(data=cust, x='Profession', hue='Gender', multiple='stack', bins=20)

plt.title('Profession by Gender')
plt.xlabel('Count')
plt.ylabel('Profession')
plt.show()
```

```python
#Histogram
plt.hist(cust['Annual Income ($)'], bins=20, color='skyblue', edgecolor='black')

plt.title('Annual Income')
plt.xlabel('Annual Income ($)')
plt.ylabel('Count')

plt.show()
```

```python
# Pie chart
plt.figure(figsize=(10, 6))
family = cust.groupby("Profession")["Spending Score (1-100)"].sum()

plt.pie(family, labels=family.index, autopct='%1.1f%%')

plt.title("Profession vs Spending score")
plt.show()
```

```python
# Point-plot
sns.pointplot(x=cust["Profession"], y=cust["Family Size"], data=cust, color = "orange")
plt.xticks(rotation=90)
plt.title("Profession vs Family Size")
plt.show()
```
