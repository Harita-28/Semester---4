```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#sklearn methods
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
temp = pd.read_csv("temperatures.csv")
temp.head()
```

```python
temp.tail()
```

```python
temp.shape
```

```python
temp.info()
```

```python
# Drop duplicates
temp = temp.drop_duplicates()
temp.shape
```

```python
# Checking for mssing values
temp.isnull().sum()
```

```python
#Plotting graphs
x = temp["YEAR"]
y = temp["ANNUAL"]
plt.plot(x,y,'o', color="orange")
plt.xlabel("Year")
plt.ylabel("Annual")
plt.title("Year vs Annual temp")
```

```python
may_temp = temp["MAY"]
jun_temp = temp["JUN"]
yr = temp["YEAR"]
plt.figure(figsize=(10, 6))
plt.plot(yr, may_temp, label="MAY")
plt.plot(yr, jun_temp, label="June")

plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.title('Comparison of May and June temp over years')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# splitting dataset
X = temp[["YEAR"]]
Y = temp["JAN"]
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=0)
X_train
```

```python
X_test
```

```python
# Load model
model= linear_model.LinearRegression()
```

```python
# Fit the model
model.fit(X_train, Y_train)
```

```python
# predicting output
y_pred = model.predict(X_test)
```

```python
# Model plot
plt.figure(figsize=(10, 6))
plt.scatter(X_train, Y_train, color="orange")
plt.plot(X_test, y_pred, color="red", linewidth=3)
plt.title("January vs Years")
plt.xlabel("Years")
plt.ylabel("January")
plt.show()
```

```python
sns.regplot(data=temp,x=X_train,y=Y_train,)
```

```python
mae = metrics.mean_absolute_error(Y_test, y_pred)
mse = metrics.mean_squared_error(Y_test, y_pred)
r2 = np.sqrt(metrics.mean_squared_error(Y_test, y_pred))
print('Mean Absolute Error:', mae)
print('Mean Square Error:', mse)
print('Root Mean Square Error:', r2)
```
